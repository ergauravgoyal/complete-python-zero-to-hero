```python
# append(item) - add item to end
fruits = ["apple", "banana"]
fruits.append("cherry")
print("append:", fruits)  # ['apple', 'banana', 'cherry']

# insert(index, item) - add item at specific position
fruits.insert(1, "mango")
print("insert:", fruits)  # ['apple', 'mango', 'banana', 'cherry']

# extend(iterable) - add multiple items
fruits.extend(["date", "elderberry"])
print("extend:", fruits)

# remove(item) - remove first occurrence (modifies in place)
fruits.remove("mango")
print("remove: mango", fruits)

# pop(index) - remove and return item at index (default: last)
last = fruits.pop()
print("pop():", last, "→", fruits)
second = fruits.pop(1)
print("pop(1):", second, "→", fruits)

# clear() - remove all items
nums = [1, 2, 3]
nums.clear()
print("clear:", nums)  # []
# index(item) - find first index of item
letters = ["a", "b", "c", "b", "a"]
print("index('b'):", letters.index("b"))  # 1

# count(item) - count occurrences
print("count('a'):", letters.count("a"))  # 2

# sort() - sort in place (ascending)
nums = [3, 1, 4, 1, 5]
nums.sort()
print("sort:", nums)  # [1, 1, 3, 4, 5]
nums.sort(reverse=True)
print("sort(reverse=True):", nums)  # [5, 4, 3, 1, 1]

# reverse() - reverse order in place
words = ["hello", "world"]
words.reverse()
print("reverse:", words)  # ['world', 'hello']

# copy() - shallow copy (new list, same items)
original = [1, 2, 3]
duplicate = original.copy()
duplicate.append(4)
print("original:", original, "| copy:", duplicate)
```
